## Часть 1. Приоритизация гипотез

Для начала импортируем библиотеки, которые понадобятся нам для работы.

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import scipy.stats as stats
import datetime as dt
```

Теперь загрузим и изучим данные из файла со сведениями о гипотезах.

```python
hypothesis_df=pd.read_csv('/datasets/hypothesis.csv')
pd.set_option('display.max_colwidth', -1)
display(hypothesis_df)
```

Переходим непосредственно к стоящим перед нами задачам

#### Применить фреймворк ICE для приоритизации гипотез. Отсортировать их по убыванию приоритета

Добавим к имеющимся в датафрейме данным столбец ICE, куда поместим вычисленные значения ICE по каждой из гипотез, после чего отсортируем данные по этому столбцу.

```python
hypothesis_df['ICE']=(hypothesis_df['Impact']*hypothesis_df['Confidence'])/hypothesis_df['Efforts']
display(hypothesis_df.sort_values(by='ICE', ascending=False))
```

Как видим, лидирует гипотеза "Запустить акцию, дающую скидку на товар в день рождения". Но общие выводы пока делать рано, перейдем к следующему пункту.

#### Применить фреймворк RICE для приоритизации гипотез. Отсортировать их по убыванию приоритета

Аналогично предыдущему шагу добавим к имеющимся в датафрейме данным столбец RICE, куда поместим вычисленные значения RICE по каждой из гипотез, после чего отсортируем данные по этому столбцу.

```python
hypothesis_df['RICE']=(hypothesis_df['Reach']*hypothesis_df['Impact']*hypothesis_df['Confidence'])/hypothesis_df['Efforts']
display(hypothesis_df.sort_values(by='RICE', ascending=False))
```

Как видим, теперь на первое место с большим отрывом вышла гипотеза "Добавить форму подписки на все основные страницы, чтобы собрать базу клиентов для email-рассылок". Перейдем к заключительному пункту, где сравним два подхода к приоритизации гипотез.

#### Указать, как изменилась приоритизация гипотез при применении RICE вместо ICE. Объяснить, почему так произошло

Учет параметра охвата гипотез (reach) существенно изменил расстановку сил между ними. Это произошло потому, что гипотеза "Запустить акцию, дающую скидку на товар в день рождения", которая занимала первое место по параметру ICE, обладает очень низким охватом. Это объясняется тем, что у интернет-магазина вполне может не быть даных о днях рождения покупателей (по крайней мере, большинства), что не позволяет отправлять им таргетированные сообщения о скидках с привязкой именно к их дню рождения. В отсутствие же такой привязки интернет-магазину остается надеяться на то, что посетитель зайдет на сайт для совершения покупки незадолго до своего дня рождения, что маловероятно в среднем. Кроме того, даже в самом лучшем случае эта акция может интересовать пользователя только один раз в году, что также не повышает охват.

Что же касается гипотезы "Добавить форму подписки на все основные страницы, чтобы собрать базу клиентов для email-рассылок", она, напротив, обладает самым выоским охватом: мы показываем эту форму всем посетителям на всех основных страницах, что позволит нам собрать большое количество email-адресов, что, в свою очередь, позволит напрямую коммуницировать с заинтересованными покупателями. Это позволяет нам вступать в контакт с пользователями, даже когда они сами не заходят на сайт. При этом затраты усилий (параметр effort) обеих гипотез одинаковы.

## Часть 2. Анализ A/B-теста

Впрочем, возможно, что эти пользователи также попадают в выбросы, исследованием которых мы займемся на следующих шагах. В таком случае будет больше оснований для того, чтобы отбросить их результаты.

Теперь перейдем непосредственно к ответам на стоящие перед нами вопросы.

#### Построить график кумулятивной выручки по группам. Сделать выводы и предположения

Начнем с того, что составим таблицу выручки по дням для группы А.

```python
ordersADaily = (orders_df[orders_df['group'] == 'A'][['date', 'transactionId', 'visitorId', 'revenue']].groupby('date', as_index=False).agg({'transactionId': pd.Series.nunique, 'revenue': 'sum'}))
ordersADaily.columns = ['date', 'ordersPerDateA', 'revenuePerDateA']
display(ordersADaily.head(10))
```

Теперь на ее основе составим таблицу кумулятивной выручки для группы А.

```python
ordersACummulative = ordersADaily.apply(
    lambda x: ordersADaily[ordersADaily['date'] <= x['date']].agg(
        {'date': 'max', 'ordersPerDateA': 'sum', 'revenuePerDateA': 'sum'}
    ),
    axis=1,
).sort_values(by=['date'])
ordersACummulative.columns = ['date','ordersCummulativeA','revenueCummulativeA',]
display(ordersACummulative.head(10))
```

Теперь проделаем все то же самое для группы В. Начнем с таблицы выручки по дням.

```python
ordersBDaily = (orders_df[orders_df['group'] == 'B'][['date', 'transactionId', 'visitorId', 'revenue']].groupby('date', as_index=False).agg({'transactionId': pd.Series.nunique, 'revenue': 'sum'}))
ordersBDaily.columns = ['date', 'ordersPerDateB', 'revenuePerDateB']
display(ordersBDaily.head(10))
```

Теперь на ее основе составим таблицу кумулятивной выручки.

```python
ordersBCummulative = ordersBDaily.apply(
    lambda x: ordersBDaily[ordersBDaily['date'] <= x['date']].agg(
        {'date': 'max', 'ordersPerDateB': 'sum', 'revenuePerDateB': 'sum'}
    ),
    axis=1,
).sort_values(by=['date'])
ordersBCummulative.columns = ['date','ordersCummulativeB','revenueCummulativeB',]
display(ordersBCummulative.head(10))
```

Теперь у нас есть все данные, нужные для построения графика кумулятивной выручки по группам.

```python
#Задаем размер поля для графиков
plt.figure(figsize=(15,5))
#Строим графики
plt.plot(ordersACummulative['date'], ordersACummulative['revenueCummulativeA'], label='A')
plt.plot(ordersBCummulative['date'], ordersBCummulative['revenueCummulativeB'], label='B')
#Работаем над внешним видом
plt.xlabel('Дата')
plt.ylabel('Выручка')
plt.title('Графики кумулятивной выручки по группам')
plt.legend()
#Выводим результат
plt.show()
```

Как видим, кумулятивная выручка группы В почти сразу вырывается вперед относительно А, а в какой-то момент разрыв становится большим. Но однозначные выводы делать пока рано: как мы видим, всплеск роста пришелся на один день, что может говорить об одном аномально большом заказе, который и внес дисбаланс. Кроме того, количество заказов в двух группах также скорее всего неодинаково, что тоже может объяснять различия в кумулятивной выручке. Более правильные выводы мы сможем сделать на следующих шагах, анализируя относительные показатели и очищенные данные. Пока что переходим к следующему шагу.

#### Построить график кумулятивного среднего чека по группам. Сделать выводы и предположения

Разделим кумулятивную выручку на кумулятивное число заказов, чтобы получить данные о среднем чеке.

```python
#Задаем размер поля для графиков
plt.figure(figsize=(15,5))
#Строим графики
plt.plot(ordersACummulative['date'], ordersACummulative['revenueCummulativeA']/ordersACummulative['ordersCummulativeA'], label='A')
plt.plot(ordersBCummulative['date'], ordersBCummulative['revenueCummulativeB']/ordersBCummulative['ordersCummulativeB'], label='B')
#Работаем над внешним видом
plt.xlabel('Дата')
plt.ylabel('Средний чек')
plt.title('Графики кумулятивного среднего чека по группам')
plt.legend()
#Выводим результат
plt.show()
```

Как видим, кумулятивный средний чек группы В сначала выше, чем у группы А, затем начинает отставать, а потом резко взлетает вверх за один день, после чего идет его падение. На этом графике необходимость очистки данных от выбросов видна еще ярче, чем на первом. Скорее всего, один или несколько аномальных заказов существенно исказили общую картину. Очисткой данных мы займемся на следующих шагах.

#### Построить график относительного изменения кумулятивного среднего чека группы B к группе A. Сделать выводы и предположения

Посмотрим на те же самые данные с другой стороны - с точки зрения изменения кумулятивного среднего чека группы В по сравнению с группой А. Начнем с того, что сведем данные в один датафрейм.

```python
mergedCumulativeRevenue = ordersACummulative.merge(ordersBCummulative, left_on='date', right_on='date', how='left')
display(mergedCumulativeRevenue.head(10))
```

Теперь переходим непосредственно к построению графика.

```python
#Задаем размер поля для графика
plt.figure(figsize=(15,5))
#Строим график отношения средних чеков
plt.plot(mergedCumulativeRevenue['date'], (mergedCumulativeRevenue['revenueCummulativeB']/mergedCumulativeRevenue['ordersCummulativeB'])/(mergedCumulativeRevenue['revenueCummulativeA']/mergedCumulativeRevenue['ordersCummulativeA'])-1)
# добавляем ось X
plt.axhline(y=0, color='black', linestyle='--')
#Работаем над внешним видом
plt.xlabel('Дата')
plt.ylabel('Относительное изменение кумулятивного среднего чека')
plt.title('График относительного изменения кумулятивного среднего чека')
#Выводим результат
plt.show()
```

Как видно из этого графика, и резкое падение среднего чека группы В относительно А, и его последующий резкий рост могут быть вызваны выбросами. На этом графике более заметно становится, что и резкое падение, которое предшествовало резкому росту, само по себе тоже может быть аномальным. Скорее всего, после очистки данных вид графика будет совсем другим.

#### Построить график кумулятивной конверсии по группам. Сделать выводы и предположения

Для построения этого графика нам впервые понадобятся данные из датафрейма с информацией о посещениях. Для начала выделим данные отдельно по группе А и по группе В. Начнем с группы A.

```python
#Выделяем группу А
visitorsADaily = visitors_df[visitors_df['group'] == 'A'][['date', 'visitors']]
#Даем понятные названия столбцам
visitorsADaily.columns = ['date', 'visitorsPerDateA']
#Посмотрим, что получилось
display(visitorsADaily.head(10))
```

Теперь делаем все то же самое для группы B.

```python
#Выделяем группу B
visitorsBDaily = visitors_df[visitors_df['group'] == 'B'][['date', 'visitors']]
#Даем понятные названия столбцам
visitorsBDaily.columns = ['date', 'visitorsPerDateB']
#Посмотрим, что получилось
display(visitorsBDaily.head(10))
```

Теперь рассчитаем кумулятивное число посетителей. Сначала для группы А.

```python
visitorsACumulative = visitorsADaily.apply(
    lambda x: visitorsADaily[visitorsADaily['date'] <= x['date']].agg(
        {'date': 'max', 'visitorsPerDateA': 'sum'}
    ),
    axis=1,
)
visitorsACumulative.columns = ['date', 'visitorsCumulativeA']
display(visitorsACumulative.head(10))
```

Теперь то же самое для группы B.

```python
visitorsBCumulative = visitorsBDaily.apply(
    lambda x: visitorsBDaily[visitorsBDaily['date'] <= x['date']].agg(
        {'date': 'max', 'visitorsPerDateB': 'sum'}
    ),
    axis=1,
)
visitorsBCumulative.columns = ['date', 'visitorsCumulativeB']
display(visitorsBCumulative.head(10))
```

Наконец, объединим все эти данные в один датафрейм с данными о заказах, полученными нами ранее.

```python
cumulative_data = (
    mergedCumulativeRevenue.merge(
        ordersADaily, left_on='date', right_on='date', how='left'
    )
    .merge(ordersBDaily, left_on='date', right_on='date', how='left')
    .merge(visitorsADaily, left_on='date', right_on='date', how='left')
    .merge(visitorsBDaily, left_on='date', right_on='date', how='left')
    .merge(visitorsACumulative, left_on='date', right_on='date', how='left')
    .merge(visitorsBCumulative, left_on='date', right_on='date', how='left')
)
display(cumulative_data.head(10))
```

Данные собраны, можно считать кумулятивную конверсию.

```python
#Считаем кумулятивную конверсию отдельно для группы A и для группы B
cumulative_data['conversionA'] = cumulative_data['ordersCummulativeA']/cumulative_data['visitorsCumulativeA']
cumulative_data['conversionB'] = cumulative_data['ordersCummulativeB']/cumulative_data['visitorsCumulativeB']
display(cumulative_data.head(10))
```

Конверсия посчитана, все готово для построения графика.

```python
#Задаем размер поля для графиков
plt.figure(figsize=(15,5))
#Строим графики
plt.plot(cumulative_data['date'], cumulative_data['conversionA'], label='A')
plt.plot(cumulative_data['date'], cumulative_data['conversionB'], label='B')
#Работаем над внешним видом
plt.xlabel('Дата')
plt.ylabel('Конверсия')
plt.title('График кумулятивной конверсии по группам')
plt.legend()
#Выводим результат
plt.show()
```

Как видим, после первоначального пика конверсия группы А далее падает и оказывается устойчиво меньше конверсии группы B. Поскольку для данного показателя влияние единичных крупных заказов существенно меньше, этот результат уже больше похож на установленную закономерность: конверсия группы В кажется больше конверсии группы А. Впрочем, эту гипотезу все равно еще предстоит проверить на очищенных данных.

#### Построить график относительного изменения кумулятивной конверсии группы B к группе A. Сделать выводы и предположения

Взглянем на те же самые данные, что и в предыдущем шаге, под другим углом - построим график относительного изменения кумулятивной конверсии группы B к группе A.

```python
#Задаем размер поля для графиков
plt.figure(figsize=(15,5))
#Строим график
plt.plot(cumulative_data['date'], cumulative_data['conversionB']/cumulative_data['conversionA']-1)
# добавляем ось X
plt.axhline(y=0, color='black', linestyle='--')
#Работаем над внешним видом
plt.xlabel('Дата')
plt.ylabel('Относительное изменение кумулятивной конверсии')
plt.title('График относительного изменения кумулятивной конверсии')
#Выводим результат
plt.show()
```

Как мы видим, не считая провала незадолго после начала теста, изменение кумулятивной конверсии группы В относительно группы А не только больше нуля, но и имеет тенденцию к росту в последние даты теста после периода падения. Таким образом, вероятность, что превышение кумулятивной конверсии группы В над группой А носит не случайный характер, довольно велика. Как видим, устойчивость результатов после начальных колебаний достаточно высока. Впрочем, окончательные выводы мы сделаем на следующих шагах.

#### Построить точечный график количества заказов по пользователям. Сделать выводы и предположения

#### Посчитать статистическую значимость различий в конверсии между группами по «сырым» данным. Сделать выводы и предположения

Начнем с того, что выделим количество заказов на каждого пользователя в разбивке по группам. Начнем с группы A.

```python
ordersByUsersA = (
    orders_df[orders_df['group'] == 'A']
    .groupby('visitorId', as_index=False)
    .agg({'transactionId': pd.Series.nunique})
)
ordersByUsersA.columns=['visitorId', 'orders']
display(ordersByUsersA.head(10))
```

Теперь то же самое для группы B.

```python
ordersByUsersB = (
    orders_df[orders_df['group'] == 'B']
    .groupby('visitorId', as_index=False)
    .agg({'transactionId': pd.Series.nunique})
)
ordersByUsersB.columns=['visitorId', 'orders']
display(ordersByUsersB.head(10))
```

Теперь создадим датафреймы, в которых каждому пользователю, сделавшему хотя бы один заказ, будет соответствовать число его заказов, а пользователям, не сделавшим ни одного заказа, будут соответствовать нули.

Теперь применим тест Манна-Уитни для определения статистической значимости различия в конверсии между двумя группами по "сырым" данным. Предварительно сформулируем нулевую и альтернативную гипотезы. Нулевая гипотеза состоит в том, что конверсия группы А и группы B одинакова. Альтернативная гипотеза состоит в том, что конверсии групп различаются.

```python
#Проверяем статистическую значимость различий в конверсии между группами
print("{0:.5f}".format(stats.mannwhitneyu(sampleA, sampleB)[1]))
#Выводим относительное различие в конверсии между группами
print("{0:.3f}".format((cumulative_data['ordersPerDateB'].sum()/cumulative_data['visitorsPerDateB'].sum())/(cumulative_data['ordersPerDateA'].sum()/cumulative_data['visitorsPerDateA'].sum())-1))
```

Итак, как мы видим, p-value составляет 0,008 - это значительно меньше 5%, что позволяет нам считать различия статистически значимыми. Это значит, что мы можем отвергнуть нулевую гипотезу и считать, что конверсии групп А и В неравны. Разница составляет 0,138 - это значит, что конверсия группы B на сырых данных на 13,8 больше конверсии группы А. На следующих шагах мы проверим, сохраняется ли эта разница на очищенных данных, но, как мы уже писали выше, можно предположить, что эта метрика более устойчива к выбросам и общая картина сохранится.

#### Посчитать статистическую значимость различий в среднем чеке заказа между группами по «сырым» данным. Сделать выводы и предположения

Перейдем к вычислению статистической значимости различий в среднем чеке заказа между группами по "сырым" данным. В данном случае все необходимые вычисления уже были произведены ранее, нам остается только вычислить различие в среднем чеке и его статистическую значимость. Снова применим тест Манна-Уитни. В данном случае наша нулевая гипотеза состоит в том, что средний чек группы А равен среднему чеку группы В. Альтернативная гипотеза состоит в том, что средний чек группы А не равен среднему чеку группы В.

```python
#Проверяем статистическую значимость различий в среднем чеке заказа между группами
print('{0:.3f}'.format(stats.mannwhitneyu(orders_df[orders_df['group']=='A']['revenue'], orders_df[orders_df['group']=='B']['revenue'])[1]))
#Выводим относительное различие в среднем чеке заказа между группами
print('{0:.3f}'.format(orders_df[orders_df['group']=='B']['revenue'].mean()/orders_df[orders_df['group']=='A']['revenue'].mean()-1))
```

Это согласуется с выводами, которые мы сделали на предыдущих шагах: на полученные данные существенное влияние оказали выбросы, что не дает нам возможность сделать однозначные выводы по ним. Это значит, что необходимо сравнивать очищенные данные, чем мы и займемся на следующих шагах.

#### Посчитать статистическую значимость различий в конверсии между группами по «очищенным» данным. Сделать выводы и предположения

Данные по конверсии уже казались достаточно достоверными, но перестраховаться никогда не будет лишним. Отбросим аномальных пользователей и вычислим статистическую значимость конверсии по очищенным данным.

Начнем с выделения множества аномальных пользователей. Как мы выяснили на предыдущих шагах, это пользователи, у которых есть либо более двух заказов, либо заказы на сумму более 28000.

```python
#Выделяем пользователей, у которых больше двух заказов
usersWithManyOrders = ordersByUsers[ordersByUsers['orders'] > 2]['visitorId']
#Выделяем пользователей с заказами дороже 28000
usersWithExpensiveOrders = orders_df[orders_df['revenue'] > 28000]['visitorId']
#Объединяем первых со вторыми и удаляем дубликаты
abnormalUsers = (
    pd.concat([usersWithManyOrders, usersWithExpensiveOrders], axis=0)
    .drop_duplicates()
    .sort_values()
)
#Посмотрим, что получилось
display(abnormalUsers.head(10))
```

Теперь модифицируем созданные нами ранее датафреймы Sample A и Sample B, исключив из них аномальных пользователей.

```python
sampleAFiltered = pd.concat([ordersByUsersA[np.logical_not(ordersByUsersA['visitorId'].isin(abnormalUsers))]['orders'],pd.Series(0, index=np.arange(visitorsADaily['visitorsPerDateA'].sum() - len(ordersByUsersA['orders'])), name='orders')],axis=0)
sampleBFiltered = pd.concat([ordersByUsersB[np.logical_not(ordersByUsersB['visitorId'].isin(abnormalUsers))]['orders'],pd.Series(0, index=np.arange(visitorsBDaily['visitorsPerDateB'].sum() - len(ordersByUsersB['orders'])), name='orders')],axis=0)
```

Наконец, вычислим статистическую значимость различий в конверсии между группами по очищенным данным. В данном случае наша нулевая гипотеза состоит в том, что конверсия группы А равна конверсии группы В. Альтернативная гипотеза состоит в том, что конверсии групп различаются.

```python
#Проверяем статистическую значимость различий в конверсии между группами
print('{0:.5f}'.format(stats.mannwhitneyu(sampleAFiltered, sampleBFiltered)[1]))
#Выводим относительное различие в конверсии между группами
print('{0:.3f}'.format(sampleBFiltered.mean()/sampleAFiltered.mean()-1))
```

Что ж, очистка данных дала свои плоды - p-value стал меньше, а относительное различие в конверсии между группами - больше. Это значит, что мы с большой уверенностью можем отвергнуть нулевую гипотезу о равнестве конверсий и придерживаться альтернативной гипотезы о том, что конверсии групп А и В неравны. #### Посчитать статистическую значимость различий в среднем чеке заказа между группами по «очищенным» данным. Сделать выводы и предположения

Итак, перед нами остался последний невыясненный вопрос - позволит ли очистка данных поднять статистическую значимость различий в среднем чеке между двумя группами до приемлемой величины? Все необходмые предварительные вычисления мы уже произвели на предыдущих шагах, осталось только непосредственно сравнить средние чеки. Наша нулевая гипотеза в данном случае состоит в том, что средний чек группы А равен среднему чеку группы В. Альтернативная гипотеза состоит в том, что средний чек группы А не равен среднему чеку группы В.

```python
#Проверяем статистическую значимость различий в среднем чеке заказа между группами
print(
    '{0:.3f}'.format(
        stats.mannwhitneyu(
            orders_df[
                np.logical_and(
                    orders_df['group'] == 'A',
                    np.logical_not(orders_df['visitorId'].isin(abnormalUsers)),
                )
            ]['revenue'],
            orders_df[
                np.logical_and(
                    orders_df['group'] == 'B',
                    np.logical_not(orders_df['visitorId'].isin(abnormalUsers)),
                )
            ]['revenue'],
        )[1]
    )
)
#Выводим относительное различие в среднем чеке заказа между группами
print(
    '{0:.3f}'.format(
        orders_df[
            np.logical_and(
                orders_df['group'] == 'B',
                np.logical_not(orders_df['visitorId'].isin(abnormalUsers)),
            )
        ]['revenue'].mean()
        / orders_df[
            np.logical_and(
                orders_df['group'] == 'A',
                np.logical_not(orders_df['visitorId'].isin(abnormalUsers)),
            )
        ]['revenue'].mean()
        - 1
    )
)
```

Что ж, результат получился несколько неожиданным: разрыв между двумя группами не только сократился, но и сменился на противоположный - теперь средний чек группы А больше среднего чека группы B. Но более удивительно то, что p-value не только не снизилось, но и возросло. Таким образом, мы не можем отвергнуть нулевую гипотезу о том, что различий в среднем чеке нет - тест говорит о том, что полученные результаты с большой вероятностью являются случайными. Вероятно, это объясняется тем, что т.к. пользователи в среднем делают по одному заказу, колебания на таком временном промежутке остаются слишком большими, чтобы выявить однозначную закономерность. А значит, достаточных оснований, чтобы придерживаться альтернативной гипотезы о неравенстве средних чеков, у нас нет.
